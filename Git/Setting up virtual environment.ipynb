# Virtual environment
- Consistency and Reproducibility
    - By using a requirements.txt file, you can ensure that anyone who sets up the virtual environment will install the exact same versions of the dependencies.
- Doesn't save disk space

# How to replicate virtual environment set up

Activate the virtual environment for the existing project and export the dependencies to a requirements.txt file:

- run in vscode terminal

```python
pip freeze > requirements.txt
```

This creates a requirements.txt file in the project folder


- When you install a package using pip, it doesn't just install the package itself; it also installs any dependencies that the package requires to function correctly. In the case of Flask, it has several dependencies, such as Werkzeug, Jinja2, itsdangerous, click, and more.

The requirements.txt file shows the version of Flask, and some other libraries that were installe from the single 'pip install Flask' command


Create a new virtual environment in the new project folder.

Activate the new virtual environment and install the dependencies from the requirements.txt file:
    - copy paste the requirements.txt file into the new project folder

```python
pip install -r requirements.txt
```
