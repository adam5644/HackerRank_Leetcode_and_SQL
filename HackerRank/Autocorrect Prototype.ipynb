```python
def f(words, queries):
    res = []
    
    # turn each word and query into counter
    from collections import Counter
    w_c = [Counter(x) for x in words]
    q_c = [Counter(x) for x in queries]
    
    for i in range(len(queries)):
        temp = []
        #print('queries[i] = ',queries[i])
        for j in range(len(words)):
            if w_c[j] == q_c[i]:
                #print('words[j] = ',words[j])
                temp.append(words[j])
        res.append(temp)
            
            
    
    return res
    
words = ['duel','speed','dule','cars']
queries = ['spede','duel']
print(f(words,queries))
```

```python
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

def get_anagrams(word, word_list):
    return sorted([w for w in word_list if are_anagrams(w, word)])

def getSearchResults(words, queries):
    return [get_anagrams(query, words) for query in queries]

# Example usage
words_example = ["duel", "speed", "dule", "cars"]
queries_example = ["spede", "duel"]
result = getSearchResults(words_example, queries_example)
result
```
