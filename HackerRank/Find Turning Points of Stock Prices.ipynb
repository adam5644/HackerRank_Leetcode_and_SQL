```python
# Provided list of prices and minimum difference n
prices = [8.1, 8.0, 6.5, 7.4, 7.9, 6.6]
n = 1.0

# Initialize counters for turning points for condition (a) and (b)
turning_points_a = 0
turning_points_b = 0
last_turning_point_value = prices[0]

# Loop over the list to find turning points
for i in range(1, len(prices) - 1):
    # Check for a turning point with no restrictions (condition a)
    if (prices[i-1] < prices[i] > prices[i+1]) or (prices[i-1] > prices[i] < prices[i+1]):
        turning_points_a += 1
        # For condition (b), we need to check if the turning point also satisfies the minimum difference
        if abs(prices[i] - last_turning_point_value) >= n:
            turning_points_b += 1
            last_turning_point_value = prices[i]

# If there's only one turning point, it automatically satisfies condition (b) since there's no previous point to compare to
if turning_points_a == 1:
    turning_points_b = 1

print(turning_points_a, turning_points_b)
```
