```python
from collections import Counter

def min_length_substrings(s, arr):
    def min_length_for_substring(target):
        target_count = Counter(target)
        current_count = Counter()
        required = len(target)
        left = 0
        min_length = float('inf')

        for right, char in enumerate(s):
            current_count[char] += 1

            if current_count[char] <= target_count[char]:
                required -= 1

            if required == 0:
                while current_count[s[left]] > target_count[s[left]]:
                    current_count[s[left]] -= 1
                    left += 1
                min_length = min(min_length, right - left + 1)

#                 # Move left pointer to find smaller window
#                 current_count[s[left]] -= 1
#                 left += 1
#                 required += 1

        return min_length if min_length != float('inf') else -1

    # For each string in arr, calculate the minimum length of the substring in s
    return [min_length_for_substring(target) for target in arr]

# Example test case
s = "111222333444"
arr = ["121", "3", "12345", "11234"]
result = min_length_substrings(s, arr)
result
```
