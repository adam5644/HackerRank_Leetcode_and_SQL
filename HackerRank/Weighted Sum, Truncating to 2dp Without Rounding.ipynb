```python
input_list = [1,2,3]

c = 0.5

def calc_s(input_list, c):
    res = 0
    for i in range(len(input_list)):
        res += (c **i * input_list[i] ) 
    return res

f_res = []
for i in range(len(input_list)):
    rotation = input_list[i:] + input_list[:i]
    f_res.append(calc_s(rotation, c))
    
def special_round(x):
    x = str(x)
    
    if '.' in x:
        part1 = x.split('.')[0]
        part2 = x.split('.')[1]
        res = part1 + '.' + (part2+'00')[:2]

    else:
        res = x + '.' + '00'
    
    return res
    
print(special_round(min(f_res)), special_round(max(f_res)))
```
